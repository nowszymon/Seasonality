```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
nfp = pd.read_excel('payrollsy.xlsx')
```

```python
nfp
```

```python
miesiace = nfp.pivot_table(index='Miesiąc',columns='Rok',values='Różnica')
```

```python
miesiace
```

```python
sns.heatmap(miesiace, linewidths=1)
plt.title('NFP - Powyżej/Poniżej prognozy')
```

```python
sns.clustermap(miesiace)
```

```python
sns.lmplot(x='Miesiąc',y='Różnica', data=nfp)
```
